```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver import ActionChains
```

```python
apsrtc_routes_df =pd.read_csv(r"C:\Users\shank\Desktop\redbus\apsrtc_route_details.csv")
apsrtc_routes_df
```

```python

driver = webdriver.Chrome()
# Initialize lists
Bus_name = []
Bus_type = []
Departure_time = []
Arrival_time = []
Ratings = []
Duration = []
Price = []
Seats_available = []
Route_link =[]
Route_name = []

def scrolling():
    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
            # Scroll down to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

            # Wait to load page
        time.sleep(1)

            # Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

for index, row in apsrtc_routes_df.iterrows():
    route_link = row['routes_link']
    route_name = row["route_names"]
    
    driver.get(route_link)
    time.sleep(5)
    
    try:
        # Click on 'View Buses' button
        buttons = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//div[@class='button']"))
        )
        buttons.click()
        time.sleep(1)
        scrolling()
    except:
        continue

    # Extract bus details
    try:
        bus_names = driver.find_elements(By.XPATH, "//div[@class='travels lh-24 f-bold d-color']")
        bus_types = driver.find_elements(By.XPATH, "//div[@class='bus-type f-12 m-top-16 l-color evBus']")
        departure_times = driver.find_elements(By.XPATH, "//*[@class='dp-time f-19 d-color f-bold']")
        durations = driver.find_elements(By.XPATH, "//div[@class='dur l-color lh-24']")
        arrival_times = driver.find_elements(By.XPATH, "//div[@class='bp-time f-19 d-color disp-Inline']")
        ratings = driver.find_elements(By.XPATH, "//div[@class='clearfix row-one']/div[@class='column-six p-right-10 w-10 fl']")
        prices = driver.find_elements(By.XPATH, "//*[@class='fare d-block']")
        seats_availables = driver.find_elements(By.XPATH, "//div[contains(@class, 'seat-left')]")
    except Exception as e:
        print(f"Error finding elements: {e}")

    # Populate lists
    for bus_name in bus_names:
         Bus_name.append(bus_name.text)
         Route_name.append(route_name)
         Route_link.append(route_link)

    for bus_type in bus_types:
         Bus_type.append(bus_type.text)

    for departure_time in departure_times:
         Departure_time.append(departure_time.text) 


    for duration in durations:
         Duration.append(duration.text)

    for arrival_time in arrival_times:
         Arrival_time.append(arrival_time.text)


    for rating in ratings:
         Ratings.append(rating.text)

    for price in prices:
         Price.append(price.text)
    
    for seats_available in seats_availables:
         Seats_available.append(seats_available.text)
    
    time.sleep(3)
    
print("Accomplished")

    # Close the WebDriver
driver.quit()
```
