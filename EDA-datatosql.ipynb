Andhra df EDA 

```python
Apsrtc_bus_details = pd.read_csv(r"C:\Users\shank\Desktop\redbus\apsrtc_bus_details.csv")
Apsrtc_bus_details
```

```python
Apsrtc_bus_details.info()
```

```python
# Remove INR from Price and convert to integer using lambda
Apsrtc_bus_details['Price'] = Apsrtc_bus_details['Price'].apply(lambda x: int(float(x.replace('INR ', '').replace(',', ''))))

#Rating conversion
Apsrtc_bus_details['Ratings'] = Apsrtc_bus_details['Ratings'].replace('New', '0')
Apsrtc_bus_details['Ratings'] = Apsrtc_bus_details['Ratings'].replace(' ', '0')

Apsrtc_bus_details['Ratings'] = Apsrtc_bus_details['Ratings'].str.split('\n').str[0].astype(float)

# Extract numeric values from Seats_available using lambda
Apsrtc_bus_details['Seats_available'] = Apsrtc_bus_details['Seats_available'].apply(lambda x: int(''.join(filter(str.isdigit, x))))

```

```python
#time convertion
keralartc_bus_details['Departure_time'] = keralartc_bus_details['Departure_time'].astype(str).str.strip()
keralartc_bus_details['Arrival_time'] = keralartc_bus_details['Arrival_time'].astype(str).str.strip()

# Convert to datetime objects first
keralartc_bus_details['Departure_time'] = pd.to_datetime(keralartc_bus_details['Departure_time'], format='%H:%M', errors='coerce').dt.time
keralartc_bus_details['Arrival_time'] = pd.to_datetime(keralartc_bus_details['Arrival_time'], format='%H:%M', errors='coerce').dt.time
```
